```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
nhl_filepath = "dataset/NHL Concussions Database_data.csv"

# Read the file into a variable nhl_data
nhl_data = pd.read_csv(nhl_filepath)
```

```python
# Print the first 5 rows of the data
nhl_data.head()
```

```python
list(nhl_data.columns)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(50,20))

# Add title
plt.title("NHL Games missed Due to Concussion in 2012-2023")

# Line chart 
sns.lineplot(data=nhl_data['Games Missed'], label="Games Missed")

# Add label for horizontal axis
plt.xlabel("Season")
```
